````python
import os
import pandas as pd
from datetime import datetime, timedelta
import yfinance as yf

# Define your stock list
stock_list = ['AAPL', 'MSFT', 'AMZN']

# Define your output directory
output_dir = "./stock_data/"

# Define the start date and fallback dates
start_date = datetime(2002, 10, 31).strftime('%Y-%m-%d')
fallback_dates = ['2006-01-31', '2008-01-31']

# Create the output directory if it doesn't exist
if not os.path.exists(output_dir):
    os.makedirs(output_dir)

# Loop through each stock ticker in the list
for stock in ticker_list:

    # Define the end date
    end_date = datetime.now().strftime('%Y-%m-%d')
    
    # Call the yfinance library to get the historical stock data
    data = yf.download(stock, start=start_date, end=end_date)
    
    # If there is no data available from the start date, use the fallback dates
    if data.loc[start_date:].empty:
        for fallback_date in fallback_dates:
            fallback_data = yf.download(stock, start=fallback_date, end=end_date)
            if not fallback_data.empty:
                data = fallback_data
                print(f"Fallback to {fallback_date} for {stock}")
                break
    
    # Select only the columns we're interested in
    data = data[['Open', 'High', 'Low', 'Close', 'Volume']]
    
    # Create the output file path
    output_file = os.path.join(output_dir, f"{stock}.csv")
    
    # Save the data to a CSV file
    data.to_csv(output_file)
    
    print(f"Saved data for {stock} to {output_file}")
````

````python
import os
import csv
import json
import pandas as pd
# Define the paths of the input files
folder_path = './stock_data/'

# Define the mapping of rating values to float values
rating_mapping = {
    "S": 4,
    "A+": 3.66,
    "A": 3.33,
    "A-": 3,
    "B+": 2.66,
    "B-":2,
    "B": 2.33,
    "C+": 1.66,
    "C-":1,
    "C": 1.33,
    "D+": 0.66,
    "D": 0.33,
    "D-":0
}
ticker_didnt_work2 = []
# Loop through each file in the folder
for filename in os.listdir(folder_path):
    if filename.endswith('.txt'):
        # Get the ticker symbol from the filename
        ticker = os.path.splitext(filename)[0]
        # Read the JSON data from the input file
        with open(os.path.join(folder_path, filename), 'r') as f:
            json_data = json.load(f)

        # Create a list of dictionaries containing the desired data from the JSON data
        rows = []
        
        for item in json_data:
          # print(item)
          # print(ticker)
          try:
            row = {
                  'Date': item.get('date',None),
                  'Rating': rating_mapping.get(item['rating'], None),
                  'RatingScore': item.get('ratingScore', None),
                  'ratingDetailsDCFScore': item.get('ratingDetailsDCFScore', None),
                  'ratingDetailsROEScore': item.get('ratingDetailsROEScore', None),
                  'ratingDetailsROAScore': item.get('ratingDetailsROAScore', None),
                  'ratingDetailsDEScore': item.get('ratingDetailsDEScore', None),
                  'ratingDetailsPEScore': item.get('ratingDetailsPEScore', None),
                  'ratingDetailsPBScore': item.get('ratingDetailsPBScore', None)
              }
            rows.append(row)
          except AttributeError:
            ticker_didnt_work2.append(ticker)
            break
        # Write the data to the output CSV file with "_2" appended to the filename
        output_file_path = os.path.join(folder_path, ticker + '_2.csv')
        with open(output_file_path, 'w', newline='') as f:
            writer = csv.DictWriter(f, fieldnames=['Date', 'Rating', 'RatingScore', 'ratingDetailsDCFScore',
                                                   'ratingDetailsROEScore', 'ratingDetailsROAScore', 'ratingDetailsDEScore',
                                                   'ratingDetailsPEScore', 'ratingDetailsPBScore'])
            writer.writeheader()
            writer.writerows(rows)

        # Merge the output CSV file with the existing CSV file with the same ticker symbol
        existing_file_path = os.path.join(folder_path, ticker + '.csv')
        if os.path.exists(existing_file_path):
            # Read in the data from the two CSV files
            df_a = pd.read_csv(existing_file_path)
            df_b = pd.read_csv(output_file_path)

            # Merge the two dataframes on the 'Date' column
            merged_df = pd.merge(df_a, df_b, on='Date', how='inner')

            # Write the merged dataframe to the newfilepath, so final result will be in ticker_2.csv
            merged_df.to_csv(output_file_path, index=False)

        # else:
        #     # Rename the output CSV file to the existing CSV file with the same ticker symbol
        #     os.rename(output_file_path, existing_file_path)
````

````python
import numpy as np
import pandas as pd
import os
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense
from tensorflow.keras.optimizers import Adam

# Define the path to the folder containing the CSV files
data_path = './stock_data'

# Define the embedding dimension
embedding_dim = 75

# Generate the initial embedding randomly
# embedding = np.random.normal(size=(7, embedding_dim))
embedding = np.random.uniform(low=-0.001, high=0.001, size=(7, embedding_dim))


# Define the learning rate and number of iterations for gradient descent
learning_rate = 0.01
num_iterations = 10

# Define the Adam optimizer with default parameters
adam = Adam()

# Loop through all the CSV files in the folder and train a separate model for each one
for filename in os.listdir(data_path):
    if filename.endswith('.csv'):
        # Load the data for the current stock
        stock_df = pd.read_csv(os.path.join(data_path, filename))
        # Split the data into training and testing sets
        train_size = int(len(stock_df) * 0.8)
        train_data = stock_df.iloc[:train_size]
        test_data = stock_df.iloc[train_size:]
        # Extract the features and target variables
        X_train = train_data.iloc[:, 2:].values
        y_train = train_data.iloc[:, 1].values
        X_test = test_data.iloc[:, 2:].values
        y_test = test_data.iloc[:, 1].values
        # Update the embedding using Adam optimizer based on the covariance matrix
        for i in range(num_iterations):
            covariance = stock_df.iloc[:, 2:].cov().values
            gradient = -2 * np.dot(covariance, embedding)
            embedding -= adam.get_updates(embedding, gradient, learning_rate)[0]
        # Define the model architecture
        model = Sequential()
        model.add(LSTM(50, input_shape=(X_train.shape[1], 1)))
        model.add(Dense(1))
        # Set the embedding layer weights to the updated embedding
        model.layers[0].set_weights([embedding.T])
        # Train the model
        model.compile(optimizer='adam', loss='mse')
        model.fit(X_train.reshape(X_train.shape[0], X_train.shape[1], 1), y_train, epochs=100, batch_size=64, verbose=0)
        # Evaluate the model on the test set
        mse = model.evaluate(X_test.reshape(X_test.shape[0], X_test.shape[1], 1), y_test, verbose=0)
        print(f'{filename}: MSE = {mse:.4f}')
````

````python
import requests
from bs4 import BeautifulSoup

def get_cik(ticker):
    url = f"https://www.sec.gov/cgi-bin/browse-edgar?CIK={ticker}&owner=exclude&action=getcompany"
    response = requests.get(url)
    soup = BeautifulSoup(response.content, "html.parser")
    cik = soup.find("span", {"class": "companyName"}).find("a").text.split("CIK")[1].strip()
    return cik
````
